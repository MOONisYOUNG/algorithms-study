### ğŸ‘‰ ê°€ëŠ¥í•œ ëª¨ë“  í•´ ì°¾ê¸°

```python
import sys

li = [[1,5,3],[2,4,7],[5,3,5]]
chk = [False]*3
m = sys.maxsize

answers = []

def backtracking(row, score):
    global m
    if row == 3: # ì¬ê·€í•¨ìˆ˜ë¥¼ ë§ˆì¹˜ëŠ” ì¡°ê±´
        answers.append(score)
        return 
    
    for i in range(0,3):
        if chk[i] == False: # ë°±íŠ¸ë˜í‚¹ì—ì„œì˜ í•œì •ì¡°ê±´
            chk[i] = True
            backtracking(row+1, score + li[row][i])
            chk[i] = False

backtracking(0,0)
print(answers)
```
